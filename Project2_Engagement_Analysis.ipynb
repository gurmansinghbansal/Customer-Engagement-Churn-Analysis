```python
def classify(row):
    if row['IsActiveMember'] == 1 and row['NumOfProducts'] > 1:
        return "Active_Engaged"
    elif row['IsActiveMember'] == 0 and row['NumOfProducts'] == 1:
        return "Inactive_LowProduct"
    elif row['IsActiveMember'] == 1:
        return "Active_LowProduct"
    else:
        return "Inactive_HighValue"

data['EngagementGroup'] = data.apply(classify, axis=1)
```

```python
data['EngagementGroup'].value_counts()
```

```python
data['BalanceSalaryRatio'] = data['Balance'] / (data['EstimatedSalary'] + 1)
```

```python
data.head()
```

```python
data.groupby('IsActiveMember')['Exited'].mean()
```

```python
data.groupby('NumOfProducts')['Exited'].mean()
```

```python
data.groupby('EngagementGroup')['Exited'].mean()
```

```python
data['BalanceGroup'] = pd.cut(
    data['BalanceSalaryRatio'],
    bins=[-1, 1, 3, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
data.groupby('BalanceGroup')['Exited'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='IsActiveMember', y='Exited', data=data)

plt.title("Churn Rate by Active Membership")
plt.xlabel("Is Active Member (0 = No, 1 = Yes)")
plt.ylabel("Churn Rate")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='NumOfProducts', y='Exited', data=data)

plt.title("Churn Rate by Number of Products")
plt.xlabel("Number of Products")
plt.ylabel("Churn Rate")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='EngagementGroup', y='Exited', data=data)

plt.title("Churn Rate by Engagement Group")
plt.xlabel("Customer Engagement Group")
plt.ylabel("Churn Rate")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='BalanceGroup', y='Exited', data=data)

plt.title("Churn Rate by Balance-Salary Group")
plt.xlabel("Balance-Salary Group")
plt.ylabel("Churn Rate")

plt.show()
```
